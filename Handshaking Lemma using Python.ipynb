```python
# Rudranarayan Jena  PRN: 20250802081

vertex = ['A', 'B', 'C', 'D']
edges = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D')]
v_pap = {v: i for i, v in enumerate(vertex)}
matrix = [[0 for _ in range(4)] for _ in range(4)]
for start,end in edges:
    row, col = v_pap[start], v_pap[end]
    matrix[row][col] = 1
    matrix[col][row] = 1
print("    A B C D")
print(" ")
for i, row in enumerate(matrix):
    print(f"{vertex[i]} | {' '.join(map(str, row))}")
# Step 1: Calculate degree of each vertex
degree = {}
for i in range(4):
    degree[vertex[i]] = sum(matrix[i])

# Step 2: Display degrees
print(f"degrees of vertices:")
for v in vertex:
    print(f"deg({v}) = {degree[v]}")

# Step 3: Verify Handshaking Lemma
total_degree = sum(degree.values())
num_edges = len(edges)
print("\nSum od degrees =", total_degree)
print("2 x Number of edges =", 2 * num_edges)
if total_degree == 2 * num_edges:
    print("Handshaking Lemma verified")
else:
    print("Handshaking Lemma NOT verified")
```
