```python
#PRN: 20250802081
import itertools

values = [True, False]

# PART 1: De Morgan's Law
print("=" * 50)
print("De Morgan's Law: ┬¼(pÔêºq) Ôëí ┬¼p Ôê¿ ┬¼q")
print("=" * 50)
print("p\t q\t ┬¼(pÔêºq)\t ┬¼pÔê¿┬¼q\t Match")
print("-" * 45)

for p, q in itertools.product(values, repeat=2):
    left = not (p and q)
    right = (not p) or (not q)
    print(f"{p}\t{q}\t{left}\t{right}\t{left == right}")

# PART 2: Implication Law
print("\n" + "=" * 50)
print("Implication Law: pÔåÆq Ôëí ┬¼p Ôê¿ q")
print("=" * 50)
print("p\t q\t pÔåÆq\t ┬¼pÔê¿q\t Match")
print("-" * 45)

for p, q in itertools.product(values, repeat=2):
    implication = (not p) or q
    disjunction = (not p) or q
    print(f"{p}\t{q}\t{implication}\t{disjunction}\t{implication == disjunction}")
```
