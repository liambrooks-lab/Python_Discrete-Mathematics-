```python
# Rudranarayan Jena   PRN: 20250802081

#Define Set S
S = {1, 2, 3}

#Define Relation R
R = {(1,1), (2,2), (3,3), (1,2), (2,1)}

#Check Reflective Property
reflexive = True
for a in S:
    if (a,a) not in R:
        reflexive = False
        print(f"Missing ({a},{a}) - Not Reflexive")

print("Reflexive:", "YES" if reflexive else "NO")

#Check Symmetric Property
symmetric = True
for (a,b) in R:
    if (b,a) not in R:
        symmetric = False
        print(f"({a},{b}) exists but ({b},{a}) missing -> Not Symmetric")
print("Symmetric:", "YES" if symmetric else "NO")

#Check Transitive Property
transitive = True
for (a,b) in R:
    for (c,d) in R:
        if b == c:
            if (a,d) not in R:
                transitive = False
                print(f"({a},{b}) and ({c},{d}) exixts but ({a},{d}) missing -> Not Transitive")

print("Transitive:", "YES" if transitive else "NO")

#Final Conclusion
if reflexive and symmetric and transitive:
    print("R is an Equivalence Relation")
else:
    print("R is NOT an Equivalence Relation")
```

```python
# Name: Rudranarayan Jena    PRN: 20250802081

vertice = ['A', 'B', 'C', 'D']
edges = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D')]

V_map = {v: i for i, v in enumerate(vertice)}

matrix = [[0 for _ in range(4)] for _ in range(4)]

for start, end in edges:
    row, col = V_map[start], V_map[end]
    matrix[row][col] = 1
    matrix[col][row] = 1

print("   A B C D")
print("_______________________________")
for i, row in enumerate(matrix):
    print(f"{vertice[i]} | {' '.join(map(str, row))}")
```
